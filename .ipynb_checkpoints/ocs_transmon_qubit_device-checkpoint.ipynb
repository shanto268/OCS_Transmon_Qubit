```python
# lambda/4 resonators
def add_cpw(total_length = '4.0mm', 
            coupling_length = '180um', 
            finishing_length = '110um',
            pos_x_i = '-30um', 
            pos_y_i = '-1024um',
            pos_x_f = '-1120um', 
            pos_y_f = '-900um',
            index = 1,
            left = True
           ):
    if left:
        otg11 = OpenToGround(design, f'open1i_{index}', options=dict(pos_x=pos_x_i,  pos_y=pos_y_i, orientation='-90',termination_gap='0um'))
        otg12 = OpenToGround(design, f'open1o_{index}', options=dict(pos_x=pos_x_f,  pos_y=pos_y_f, orientation='180',termination_gap='0um'))
    else:
        otg11 = OpenToGround(design, f'open1i_{index}', options=dict(pos_x=pos_x_i,  pos_y=pos_y_i, orientation='-90',termination_gap='0um'))
        otg12 = OpenToGround(design, f'open1o_{index}', options=dict(pos_x=pos_x_f,  pos_y=pos_y_f, orientation='0',termination_gap='0um'))
    
    
    
    options = Dict(pin_inputs=Dict(start_pin=Dict(
                                                component=f'open1o_{index}',
                                                pin='open'),
                                            end_pin=Dict(
                                                component=f'open1i_{index}',
                                                pin='open')
                                            ),
                    lead=Dict(
                            start_straight=coupling_length,
                            end_straight=finishing_length),
                    meander=Dict(
                            spacing = '78um'),
                          total_length=total_length,
                          fillet='38.5um')


    cpw = RouteMeander(design, f'cpw_{index}', options=options)

    gui.rebuild()
    gui.autoscale()
    gui.screenshot()
    
    return cpw
```

```python
design.overwrite_enabled = True
def add_diss(index=2, 
             pos_x=f'{-1420+179}um',
             pos_y= 0,
             
             cap_width = '10um',
            cap_gap = '6um',
            cap_gap_ground = '28um',
            finger_length = '75um',
            finger_count = '9',
            termination_cap_width = '5um',
            termination_cap_gap = '7um',
            termination_cap_gap_ground = '15um',
            termination_finger_length = '8um',
            termination_finger_count = '3',
            cross_width = '14um',
            cross_length = '70um',
            cross_gap = '30um',
            left = True
             ):

    options_d = dict(
        make_fl = False,
        pos_x = pos_x, 
        pos_y = pos_y, 
        orientation = 90 if left else 270,
        pin_orientation = -90 if left else 90,
        flux_orientation = -180 if left else 0,

        cap_distance = '1um',
        cap_width = cap_width,
        cap_gap = cap_gap,
        cap_gap_ground = cap_gap_ground,
        finger_length = finger_length,
        finger_count = finger_count,
        termination_cap_width = termination_cap_width,
        termination_cap_gap = termination_cap_gap,
        termination_cap_gap_ground = termination_cap_gap_ground,
        termination_finger_length = termination_finger_length,
        termination_finger_count = termination_finger_count,
        cross_width = cross_width,
        cross_length = cross_length,
        cross_gap = cross_gap,
        
        hfss_inductance = '3.6nH',
        chip='main',
        # Adding 3 connectors 
        connection_pads=dict(
            # ground_pin  = dict(connector_location = '0', connector_type = '0', claw_width = '19.1um', claw_length = '35.9um', ground_spacing = '5um')
        ),

    )

    q_diss = myTransmonNewCrosswTeeth(design, f'diss_{index}', 

                                       options=options_d)
    gui.rebuild()
    return q_diss
```
